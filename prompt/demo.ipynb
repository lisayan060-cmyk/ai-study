# 提示词工程 (Prompt Engineering)

学习如何编写有效的提示词，让AI模型返回更准确的回复。

```python
from 提示词工程 import generate_responses
```

## 1. 定义清楚提示词

给模型提供上下文信息，让它根据上下文回答问题，而不是依赖自身知识。

```python
# 定义结构，定义变量

instruction = """
根据下面的上下文回答问题。保持答案简短且准确。如果不确定答案，请回答"不确定答案"。
"""

context = """
Teplizumab起源于一个位于新泽西的药品公司，名为Ortho Pharmaceutical。\
在那里，科学家们生成了一种早期版本的抗体，被称为OKT3。最初这种分子是从小鼠中提取的。\
能够结合到T细胞的表面，并限制它们的细胞杀伤潜力。在1986年，它被批准用于帮助预防肾脏移植后的\
器官排斥，成为首个被允许用于人类的治疗性抗体。
"""

query = """
OKT4最初是从什么来源提取的？
"""

# 用f-string将三部分组合成完整的prompt
prompt = f"""
{instruction}

### 上下文
{context}

### 问题：
{query}
"""

response = generate_responses(prompt)
print(response)
```

```python
# 没有上下文的提示词 — 模型依赖自身知识回答，可能不准确
prompt = """问题：OKT3最初是从什么来源提取的？"""

response = generate_responses(prompt)
print(response)
```

## 2. 指定输出格式

通过指定输出格式（如JSON、列表），控制模型的回复结构。

```python
# 添加格式

instruction = """
根据下面的上下文回答问题。保持答案简短且准确。如果不确定答案，请回答"不确定答案"。

以Json格式输出：
{"[具体问题]":"[答案]"},
"""

context = """
Teplizumab起源于一个位于新泽西的药品公司，名为Ortho Pharmaceutical。\
在那里，科学家们生成了一种早期版本的抗体，被称为OKT3。最初这种分子是从小鼠中提取的，\
能够结合到T细胞的表面，并限制它们的细胞杀伤潜力。在1986年，它被批准用于帮助预防肾脏移植后的\
器官排斥，成为首个被允许用于人类的治疗性抗体。
"""

query = """
OKT3最初是从什么来源提取的？
"""

prompt = f"""
{instruction}

### 上下文
{context}

### 问题：
{query}
"""

response = generate_responses(prompt)
print(response)
```

## 4. Few-shot 示例

```python
# Few-shot: 提供示例让模型学习情感分类的模式
prompt = """
判断以下评论的情感是"正面"还是"负面"。
评论：娃要开学啦
情感：
"""

# 评论：这家餐厅的菜太好吃了！
# 情感：正面

# 评论：服务态度很差，等了一个小时。
# 情感：负面

# 评论：环境优美，价格合理，下次还来。
# 情感：正面

# 评论：食物又冷又难吃，非常失望。
# 情感：
# """

response = generate_responses(prompt)
print(response)
```

## 5. Chain-of-Thought 思维链

通过提供带有推理步骤的示例，引导模型逐步思考，而不是直接给出答案。

```python
prompt = """
这组数中的奇数加起来是偶数：4、8、9、15、12、2、1。
A：将所有奇数相加（9、15、1）得到25。答案为False。
这组数中的奇数加起来是偶数：17、10、19、4、8、12、24。
A：将所有奇数相加（17、19）得到36。答案为True。
这组数中的奇数加起来是偶数：16、11、14、4、8、13、24。
A：将所有奇数相加（11、13）得到24。答案为True。
这组数中的奇数加起来是偶数：17、9、10、12、13、4、2。
A：将所有奇数相加（17、9、13）得到39。答案为False。
这组数中的奇数加起来是偶数：15、32、5、13、82、7、1。
A：
"""

response = generate_responses(prompt)
print(response)
```

```python
# Zero-shot CoT：不需要提供示例，只需加上"让我们一步一步地思考"
prompt = """
我去市场买了10个苹果。我给了邻居2个，给了修理工2个。
然后我又买了5个苹果，吃了1个。我还剩多少个苹果？
让我们一步一步地思考。
"""

response = generate_responses(prompt)
print(response)
```

## 6. Tree-of-Thought 思维树

让多个"专家"从不同角度同时推理，互相验证，发现错误时及时退出，最终得到更可靠的答案。

```python
# Tree-of-Thought：多个专家从不同角度推理，互相验证
prompt = """
假设三位不同的专家来回答这个问题。
每位专家都写下他们思考的第一步，然后与大家分享。
然后所有专家写下下一步并分享。以此类推。
如果任何专家在任何时候意识到自己的推理是错的，那他就退出。

问题：小明有3本书，妈妈给了他5本，他送给同学2本后又买了4本，
后来丢了1本，爸爸又给了他和妹妹一共6本（平分）。
小明现在有几本书？
"""

response = generate_responses(prompt)
print(response)
```
